```python
import pandas as pd
```

```python
data=pd.read_csv('data_correc.csv')
```

```python
data[data['Country Code']=='IND']
```

```python
data2= data.drop(['Indicator Code','Indicator Name','Country Name','Country Name'],axis=1)
```

```python
list_of_columns=[]
for el in data2.columns:
    print(el)
    list_of_columns.append(el)
```

```python
list_of_columns.remove('Country Code')
for el in list_of_columns:
    data[el]=data[el].fillna(0)
```

```python
India_rank=[]
year=[]
for el in list_of_columns:
    data3=data2[['Country Code',el]]
    data3=data3.sort_values(by=el,ascending=False)
    an=data3['Country Code']
    an=an.to_list()
    ans=an.index('IND')
    India_rank.append(ans)
    year.append(el)
    print(ans, el)
India_rank.pop()
year.pop()
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
World_rank=[]
year=[]
for el in list_of_columns:
    data3=data2[['Country Code',el]]
    data3=data3.sort_values(by=el,ascending=False)
    an=data3['Country Code']
    an=an.to_list()
    ans=an.index('WLD')
    World_rank.append(ans)
    year.append(el)
    print(ans, el)
World_rank.pop()
year.pop()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x=year,y=India_rank,marker='o',label='India_rank')
sns.lineplot(x=year,y=World_rank,marker='*', label='World_Rank')
plt.xticks(rotation=90)
plt.legend()
plt.title('India rank according to income per person')
plt.xlabel('Year')
plt.ylabel('rank')
plt.show()
```
